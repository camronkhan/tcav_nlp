```python
import pickle
import matplotlib.pyplot as plt
import numpy as np
```

```python
def plot_graphs(scores):
    fig, axes = plt.subplots(len(scores.keys()), 1, figsize=(30, 30))

    for i, (key, val) in enumerate(scores.items()):
        labs = [key for key, val in scores[key].items()]
        vals = [val for key, val in scores[key].items()]

        axes[i].bar(labs,
                    vals,
                    alpha=0.3,
                    color='b',
                    label='TCAV',
                    width=0.2)

        axes[i].set_title(f'TCAV for "{key}" concept', size=25)
        axes[i].set_xlabel('Labels', size=25)
        axes[i].set_ylabel('TCAV', size=25)
        axes[i].set_xticklabels(labs, rotation='vertical')
        axes[i].legend(prop={'size': 15})
        axes[i].tick_params(labelsize=20)

    fig.tight_layout()
    plt.show()
```

Here we test three concepts "Москва", "Ученые" and "Концерт" using trigrams. See examples below.

```python
concepts = pickle.load(open('experiments/test/concepts.pkl', 'rb'))
```

```python
for key in concepts.keys():
    print(f'10 positive examples for "{key}" concept:\n')
    print('\n'.join(concepts[key]['pos'][:10]))
    print(f'\n10 negative examples for "{key}" concept:\n')
    print('\n'.join(concepts[key]['neg'][:10]))
    print('-' * 50, '\n')
```

## Deep Averaging Network

```python
dan_scores = pickle.load(open('experiments/dan/scores.pkl', 'rb'))
```

```python
plot_graphs(dan_scores)
```
